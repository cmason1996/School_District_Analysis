```python
# new dataframe

sales_df = pd.DataFrame({"item": ["tomatoes", "onions",
                                  
                                  "potatoes", "tomatoes", "mushrooms"],
                         
                "quantity": [15, 26, 10, 12, 2],
                         
                "total_price": [50.25, 39.52, 50.01, 40.20, 8.46]})

sales_df.head()
```

```python
# filter the dataframe by values greater then 10

sales_df["quantity"] > 10
```

```python
# filter the data so only quantity values greater than 10 are displayed

filter = sales_df["quantity"] > 10

sales_df.loc[filter]
```

```python
# combining the two lines in the previous cell

sales_df.loc[sales_df["quantity"] >10]
```

```python
# combining filters udsing &

sales_df.loc[(sales_df["quantity"] > 10) &
            (sales_df["total_price"] <45)] 

# code returns rows where both statements are true
```

```python
# combining the filters using | 'or'

sales_df.loc[(sales_df["quantity"] > 10) |
            (sales_df["total_price"] <45)] 

# code returns rows that have either one of both statements true
```

```python
# filtering using min() to find row with the lowest value

min_price = sales_df["total_price"].min()

min_price_row = sales_df.loc[sales_df["total_price"] == min_price]

min_price_row
```

```python
# filtering using max() to find the row with the highest value

max_price = sales_df["total_price"].max()

max_price_row = sales_df.loc[sales_df["total_price"] == max_price]

max_price_row
```

```python
# filtering using the mean() function

sales_df.loc[sales_df["item"] == "tomatoes", ["total_price"]].mean()
```

```python
# filtering by column using iloc[]

sales_df.iloc[0:3, [0,2]] 
```

```python
# filtering by column usinf loc[]

sales_df.loc[sales_df["item"] == "tomatoes", ["quantity", "total_price"]]

#filtering by the price and quantity columns for any rows that contain tomatoes
```

```python
# filtering by column without specifying a row

sales_df.loc[:, ["item", "total_price"]] 

# filters all rows to include only the item and price columns
```

```python
# sorting values by column, values will appear in ascending order

sales_df.sort_values("total_price")

# or 
# sales_df.sort_values("quantity") 
# sales_df.sort_values("item")
```

```python
# sorting values by column in descending order

sales_df.sort_values("total_price", ascending=False) 
```

```python
# to sort by multiple columns

sales_df.sort_values(["total_price", "item"], ascending=False) 
```
